```python
import pandas as pd
import numpy as np

# 1. Carregue o CSV e exiba as 5 primeiras linhas

df = pd.read_csv("amazon_sales_dataset.csv")
display(df.head(5))
```

```python
#2. Quantos pedidos existem por categoria de produto? Ordene do maior para o menor
ordenar = df["product_category"].value_counts()
display(ordenar)
```

```python
#3. Filtre apenas os pedidos com rating acima de 4.0 e desconto maior que 20%
ratingdesconto = df[(df['rating']> 4) & (df['discount_percent'] > 20)]
display(ratingdesconto)
```

```python
# 4. Crie uma coluna 'discount_value' que mostre o valor do desconto em dinheiro (price - discounted_price)
df["Discount_value"] = df["discounted_price"] - df["price"]
display(df)
```

```python
# 5. Qual a receita total por regiÃ£o? Qual regiÃ£o faturou mais?
lugar = df.groupby("customer_region")["total_revenue"].sum().sort_values(ascending=False)
display(lugar)
```

```python
# 6. Crie uma coluna 'performance' classificando os produtos em 'Excelente', 'Bom' ou 'Ruim' pelo rating
df["Performance"] = np.where(df["rating"]>= 4.5, "Execpcional",
                             np.where(df["rating"] < 3.5, "ruim", "bom"))
display(df)
```

```python
# 7. Qual o mÃ©todo de pagamento mais usado?

comparador = df["payment_method"].value_counts()
display(comparador)
```

```python
# 8. Filtre os produtos da categoria 'Books' com receita acima da mÃ©dia geral
media = df["total_revenue"].mean()
df["Media"] = media
filtro = df[(df["product_category"] == "Books") & (df["total_revenue"] > media)]
display(filtro)
```
